## Adjusting image brightness and contrast

```python
def adjust_brightness_contrast_gray(img, alpha=1.0, beta=0):
    # Apply brightness and contrast adjustment to the image
    adjusted_img = np.clip(alpha * img + beta, 0, 255).astype(np.uint8)

    return adjusted_img

```

```python
adj_img1 = adjust_brightness_contrast_gray(gray_img1, alpha=1.5, beta=-70)
```

```python
plt.imshow(adj_img1)
plt.axis('off')  # Hide axis
plt.show()
```

## Normalizing the images

```python
def normalize_image(image):
    # Convert image to floating point format
    normalized_image = image.astype('float32')
    
    # Normalize pixel values to the range [0, 1]
    normalized_image /= 255.0
    
    return normalized_image

```

```python
# Normalize the image
normalized_img1 = normalize_image(gray_img1)
```

```python
plt.imshow(normalized_img1)
plt.axis('off')  # Hide axis
plt.show()
```

```python
# Normalize the image
normalized_img2 = normalize_image(gray_img2)
plt.imshow(normalized_img2)
plt.axis('off')  # Hide axis
plt.show()
```
